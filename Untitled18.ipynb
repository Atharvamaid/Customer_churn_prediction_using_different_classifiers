```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/customer_churn_datasets.csv")
```

```python
df.hist()
```

```python
df.drop(columns='customerID', inplace=True)
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.hist(column='tenure')
```

```python
#data preprocessing

df1.replace("No internet service", "No", inplace=True)
df1.replace("No phone service", "No", inplace=True)
df1.head()
```

```python
#yes = 1
#no = 0

yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_columns:
  df1[col].replace({'Yes':1, 'No':0}, inplace=True)
```

```python
df1.head()
```

```python
df1.gender.replace({'Female':1,"Male":0}, inplace=True)
```

```python
X = pd.get_dummies(data=df1, columns=['InternetService', 'Contract', 'PaymentMethod'])
X.sample(3)
```

```python
from sklearn.preprocessing import MinMaxScaler

cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
scaler = MinMaxScaler()
X[cols_to_scale] = scaler.fit_transform(X[cols_to_scale])
```

```python
X.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split

features = X.drop('Churn', axis='columns')
labels = X['Churn']


X_train , X_test, Y_train, Y_test = train_test_split(features, labels, test_size=0.15)
```

```python
print(X_train.shape, Y_train.shape)
```

```python
#classification using different models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

accuracies = {}

def train_models(models, X_train, Y_train, X_test, Y_test):
  for i,j in models.items():
    model = j()
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    print(f'The accuracy for {i} is {accuracy_score(Y_test, y_pred)}')
    accuracies[i] = accuracy_score(Y_test, y_pred)*100

models = {'LogisticRegression' : LogisticRegression, 
         'GaussianNB' : GaussianNB,
         'KNeighborsClassifier' : KNeighborsClassifier,
         'DecisionTreeClassifier': DecisionTreeClassifier,
         'RandomForestClassifier' : RandomForestClassifier,
          'SVC': SVC}
train_models(models, X_train, Y_train, X_test, Y_test)
```

```python
model = list(accuracies.keys())
accuracy = list(accuracies.values())

plt.figure(figsize=(14,5))
plt.bar(model, accuracy, width=0.8)
```
